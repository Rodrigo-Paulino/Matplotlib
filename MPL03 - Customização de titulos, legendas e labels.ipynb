```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = []
y = []
y2 = []

print(x)
print(y)
```

```python
for num in range(-200,201,10):
    x.append(num)
    y.append(num)
    y2.append(2 * num)
    
    
print(x)
print(y)
```

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue")
plt.plot(x,y2, c = "green")
```

# Adicionando descrição aos Eixos

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue")
plt.plot(x,y2, c = "green")

#Primeira descrição 
plt.xlabel("Eixo X")

# 'xlabel' é para eixo X
```

# Formatando a fonte descritiva dos eixos

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue")
plt.plot(x,y2, c = "green")

#Formatando
plt.xlabel("Espaço para o Eixo X", color = "red", fontsize = 15)

# 'xlabel' é para eixo X
# color = "cor"
# fontsize = Tamanho do Texto
```

```python
#Alinhamento dos textos descritivos
```

```python
#Espaço livre a partir do ponto médio do eixo

plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue")
plt.plot(x,y2, c = "green")

#Alinhando a partir de 0

plt.xlabel("Espaço para o Eixo X", color = "red", fontsize = 15, horizontalalignment = "right")


# 'xlabel' é para eixo X
# color = "cor"
# fontsize = Tamanho do Texto
# horizontalalignment = "right, or center, or left"
```

# Y label

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue")
plt.plot(x,y2, c = "green")

#Realizando formatações no eixo Y
plt.xlabel("Espaço para o Eixo X", color = "red", fontsize = 15, horizontalalignment = "center")
plt.ylabel('Espaço para o Eixo Y', color='blue', fontsize=15, horizontalalignment='center')
```

# Alinhamento vertical 
# Alinhamos o eixo Y verticalmente atraves da opção horizontalalignment

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue")
plt.plot(x,y2, c = "green")

#Alinhando verticalmente

plt.xlabel("Espaço para o Eixo X", color = "red", fontsize = 15, horizontalalignment = "right")
plt.ylabel("Espaço para o Eixo y", color = "blue", fontsize = 15, horizontalalignment = "right") # Ele alinha do centro pra baixo
```

# Adicionando Título

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue")
plt.plot(x,y2, c = "green")

plt.xlabel("Espaço para o Eixo X", color = "red", fontsize = 15, horizontalalignment = "center")
plt.ylabel("Espaço para o Eixo y", color = "blue", fontsize = 15, horizontalalignment = "center")

#Title
plt.title("Titulo do Grafico", color = "gray", fontsize = 25)

# 'xlabel' = Nome do eixo X
# color = "cor"
# fontsize = Tamanho do Texto
# horizontalalignment = "right, or center, or left"
# plt.title = Nome do Título
```

# Texto Matamático

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue")
plt.plot(x,y2, c = "green")

plt.xlabel("Espaço para o Eixo X", color = "red", fontsize = 15, horizontalalignment = "center")
plt.ylabel("Espaço para o Eixo y", color = "blue", fontsize = 15, horizontalalignment = "center")


# verticalalignment = "bottom"
plt.title(r"Titulo do Grafico $\frac{5}{4}$", color = "gray", fontsize = 25 )

# 'r' abre uma janela de programação antes do ""
# dentro de $...$ pode colocar o número fracionário que quiser 
# \frac{5}{4} e a fração que quiser 
```

# LABELS ou legendas

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue",
        label = "Primeiro")
plt.plot(x,y2, c = "green",
        label = "Segundo")

plt.xlabel("Espaço para o Eixo X", color = "red", fontsize = 15, horizontalalignment = "center")
plt.ylabel("Espaço para o Eixo y", color = "blue", fontsize = 15, horizontalalignment = "center")

plt.title(r"Titulo do Grafico $\frac{5}{4}$", color = "gray", fontsize = 25 )

#Inserindo legenda
plt.legend()

# 'xlabel' = Nome do eixo X
# color = "cor"
# fontsize = Tamanho do Texto
# horizontalalignment = "right, or center, or left"
# plt.title = Nome do Título
# plt.legend() = Legenda
```

# Customizando os eixos (Tickers)

```python
plt.plot(x,y, c = "red", ls = "-", lw = 2, marker = "o", ms = 10, fillstyle = "top", mfc = "blue",
        label = "Primeiro")
plt.plot(x,y2, c = "green",
        label = "Segundo")

plt.xlabel("Espaço para o Eixo X", color = "red", fontsize = 15, horizontalalignment = "center")
plt.ylabel("Espaço para o Eixo y", color = "blue", fontsize = 15, horizontalalignment = "center")

plt.title("Titulo do Grafico", color = "gray", fontsize = 25 )
plt.legend(loc = "lower right")

#Iniciando a customização dos eixos

eixoX = ["Janeiro","Abril","Julho"]

plt.xticks([-200,0,200],eixoX)

# 'xlabel' = Nome do eixo X
# color = "cor"
# fontsize = Tamanho do Texto
# horizontalalignment = "right, or center, or left"
# plt.title = Nome do Título
# plt.legend() = Legenda
# plt.xticks = Substitui os ticks por outra lista, MAS TEM QUE ESPESPECIFICAR O 'DÊ' E O 'PARA'. 
```

# Customizando o Eixo X e Y

```python
plt.plot(x,y, c = "red", ls = "-", label = "Produto A")
plt.plot(x,y2, c = "green", ls = ":", label = "Produto B")

plt.xlabel("Meses", color = "red", fontsize = 15, horizontalalignment = "center")
plt.ylabel("Classificação", color = "blue", fontsize = 15, horizontalalignment = "center")

plt.legend(loc = "lower right")

#Customização de eixos X e Y

plt.title("Produtos em 2018", color = "gray", fontsize = 25 )

eixoX = ["Janeiro","Abril","Julho"]
eixoY = ["Fraco","Regular","Neutro","Bom","Excelente"]

plt.xticks([-200,0,200],eixoX)
plt.yticks([-400,-100,0,100,400],eixoY)
```

# Rotacionando Ticks

```python
eixoX = ["01/01/2018","01/04/2018","30/06/2018","01/07/2018","01/10/2018"]
eixoY = ["Fraco","Regular","Neutro","Bom","Excelente"]

plt.plot(x,y, c = "red", ls = "-", label = "Produto A")
plt.plot(x,y2, c = "green", ls = ":", label = "Produto B")

plt.xlabel("Datas", color = "red", fontsize = 15, horizontalalignment = "center")
plt.ylabel("Classificação", color = "blue", fontsize = 15, horizontalalignment = "center")

plt.title("Produtos", color = "gray", fontsize = 25 )

plt.xticks([-200,-100,0,100,200],eixoX)
plt.yticks([-400,-100,0,100,400],eixoY)

plt.legend(loc = "lower right") 
```

```python
eixoX = ["01/01/2018","01/04/2018","30/06/2018","01/07/2018","01/10/2018"]
eixoY = ["Fraco","Regular","Neutro","Bom","Excelente"]

plt.plot(x,y, c = "red", ls = "-", label = "Produto A")
plt.plot(x,y2, c = "green", ls = ":", label = "Produto B")

plt.ylabel("Classificação", color = "blue", fontsize = 15, horizontalalignment = "center")

plt.title("Produtos", color = "gray", fontsize = 25 )

plt.xticks([-200,-100,0,100,200],eixoX, rotation = 30)
plt.yticks([-400,-100,0,100,400],eixoY)

plt.legend(loc = "lower right")

# rotation = Usar de 30 a 50 graus dependendo da quantidade de dados. 
# Se for necessário retacionar 90 graus, melhor rotacionar todo o gráfico.
```
